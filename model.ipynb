```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
from datetime import datetime


```

```python
df.isnull().sum()
```

```python
features = ['discount_applied', 'unit_price', 'product_stock', 'product_category', 'day_of_week', 'season']
df=df[features]
```

```python
# Step 1: Load and Preprocess the Data
def preprocess_data(df, product_id):
    # Filter for the selected product
    product_data = df[df['product_id'] == product_id]

    # Convert transaction_date to datetime format
    product_data['transaction_date'] = pd.to_datetime(product_data['transaction_date'])

    # Set transaction_date as the index
    product_data.set_index('transaction_date', inplace=True)

    # Sort data by date
    product_data = product_data.sort_index()

    # Select relevant columns: quantity (target), unit_price, discount_applied, product_stock, day_of_week, season
    features = ['discount_applied', 'unit_price', 'product_stock', 'day_of_week', 'season']
    product_data = product_data[['quantity'] + features]

    # Convert categorical features (day_of_week, season) into dummy/one-hot encoded variables
    product_data = pd.get_dummies(product_data, columns=['day_of_week', 'season'], drop_first=True)

    return product_data
```

```python
# Step 2: Train-Test Split
def train_test_split(product_data, test_size=0.2):
    # Split data into train and test sets (80% train, 20% test)
    split_index = int(len(product_data) * (1 - test_size))
    train_data = product_data.iloc[:split_index]
    test_data = product_data.iloc[split_index:]

    return train_data, test_data


# Step 3: Build the ARIMA Model
def build_arima_model(train_data, exog_features):
    # Fit ARIMA model (order (p, d, q)) with exogenous features
    model = ARIMA(train_data['quantity'], order=(5, 1, 0), exog=train_data[exog_features])
    model_fit = model.fit()

    return model_fit

```

```python
# Step 4: Forecast Sales
def forecast_sales(model_fit, steps, exog_future):
    # Forecast sales for the future 'steps' time steps with exogenous variables
    forecast = model_fit.forecast(steps=steps, exog=exog_future)
    return forecast

# Step 5: Evaluate the Model
def evaluate_model(test_data, predicted_sales):
    # Compare predicted sales with actual sales in the test set
    mae = mean_absolute_error(test_data['quantity'], predicted_sales)
    print(f"Mean Absolute Error (MAE): {mae}")

    # Plot actual vs predicted sales
    plt.figure(figsize=(10,6))
    plt.plot(test_data.index, test_data['quantity'], label="Actual Sales")
    plt.plot(test_data.index, predicted_sales, label="Predicted Sales", linestyle='--')
    plt.title("Actual vs Predicted Sales")
    plt.xlabel("Date")
    plt.ylabel("Sales Quantity")
    plt.legend()
    plt.show()
```
